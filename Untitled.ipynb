```python
#ASSIGNMENT One
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr


#varibility of rainfall for the flood month Odawna
NC = xr.open_dataset('GPM_IMERG_kumasi.nc')
Odawna = NC.precipitationCal.sel(lon = -0.2172, lat = 5.56, method = 'nearest')
Odawna_June = Odawna.sel(time =slice('2019-06-01', '2019-06-30'))
Odawna_anon = (Odawna_June-np.mean(Odawna_June))
#percentage contribution of 9 sept, 2007 rainfall
per_cont = (Odawna_June.sel(time = '2019-06-20') /Odawna_June.sum())*100

# #Mallam_Juntion
Mallam_Juntion = NC.precipitationCal.sel(lon = -0.2850, lat = 5.57, method = 'nearest')
Mallam_Juntion_Oct = Mallam_Juntion.sel(time =slice('2020-10-01', '2020-10-31'))
Mallam_Juntion_anon = (Mallam_Juntion_Oct-np.mean(Mallam_Juntion_Oct))
#percentage contribution of 9 sept, 2007 rainfall
Dan_per_cont = (Mallam_Juntion_Oct.sel(time = '2020-10-25') /Mallam_Juntion_Oct.sum())*100

# #kwame nkrumah Ofankor
Ofankor = NC.precipitationCal.sel(lon = -0.2654, lat = 5.6634, method = 'nearest')
Ofankor_june = Ofankor.sel(time =slice('2021-06-01', '2021-06-30'))
Ofankor_anon = (Ofankor_june-np.mean(Ofankor_june))
#percentage contribution of 9 sept, 2007 rainfall
Ofankor_per_cont = (Ofankor_june.sel(time = '2021-06-03') /Ofankor_june.sum())*100


# #Santa Maria
# Santa = NC.precip.sel(longitude = -0.264, latitude = 5.604, method = 'nearest')
# santa_sept = Ofankor.sel(time =slice('2014-09-01', '2014-09-30'))
# santa_anon = (santa_sept-np.mean(santa_sept))
# #percentage contribution of 9 sept, 2007 rainfall
# santa_per_cont = (santa_sept.sel(time = '2014-09-26') /santa_sept.sum())*100



fig, ax = plt.subplots(2,2, figsize = (40, 40))
fig.subplots_adjust(bottom=0.5, left = 0.4)
#Odawna
ax[0,0].plot(Odawna_June, color = 'k', lw = 3, marker = 'o', label = 'Rainfall amount')
ax[0,0].plot(Odawna_anon, color = 'r', lw = 3, marker = 'o', label = 'Ranifall anomaly')
ax[0,0].set_ylabel('Rainfall (mmd$^{-1}$)', fontsize =18)
ax[0,0].set_xlabel('Days', fontsize =18)
ax[0,0].set_title('Odawna, June, 2019', size =20, weight ='bold')
ax[0,0].axhline(y = 10, color ='k', label = '75th Percentile' , linestyle = '-', lw =3)
ax[0,0].axhline(y = 25, color ='r', label = '95th Percentile', linestyle = '--', lw =3)
ax[0,0].set_xticks(range(1, 30, 3))
ax[0,0].set_xticklabels(['3','6', '9','12', '15', '18', '21','24', '27', '30'], size =20)
ax[0,0].set_yticks(range(-10, 60, 10))
ax[0,0].set_yticklabels(['-10','0','10', '20', '30', '40', '50'], size =20)
ax[0,0].annotate('*', xy =(21.5, 31.7), size =40, color = 'b')
ax[0,0].annotate('12.03%', xy =(20, 36.7), size =30, color = 'k')
ax[0,0].legend(loc = 2, fontsize =15)
#ax[0,0].legend(loc =1, bbox_to_anchor=(0.50, 1.0), fontsize = 15)


# #Mallam_Juntion
ax[0,1].plot(Mallam_Juntion_Oct, color = 'k', lw = 3, marker = 'o', label = 'Rainfall amount')
ax[0,1].plot(Mallam_Juntion_anon, color = 'r', lw = 3, marker = 'o', label = 'Ranifall anomaly')
ax[0,1].set_ylabel('Rainfall (mmd$^{-1}$)', fontsize =18)
ax[0,1].set_xlabel('Days', fontsize =18)
ax[0,1].set_title('Mallam Junction, October, 2020', size =20, weight ='bold')
ax[0,1].axhline(y = 10, color ='k', label = '75th Percentile' , linestyle = '-', lw =3)
ax[0,1].axhline(y = 25, color ='r', label = '95th Percentile', linestyle = '--', lw =3)
ax[0,1].set_xticks(range(1, 30, 3))
ax[0,1].set_xticklabels(['3','6', '9','12', '15', '18', '21','24', '27', '30'], size =20)
ax[0,1].set_yticks(range(-10, 60, 10))
ax[0,1].set_yticklabels(['-10','0', '10',  '20','30', '40',  '50' ], size =20)
ax[0,1].annotate('*', xy =(23.6, 30.7), size =40, color = 'b')
ax[0,1].annotate('22.56%', xy =(25, 32), size =30, color = 'k')
#ax[0,1].legend(loc =1, bbox_to_anchor=(0.50, 1.0), fontsize = 15)
ax[0,1].legend(loc = 2, fontsize =15)
#ax[0,1].legend(title = 'Legend', loc=3, bbox_to_anchor=(1, -0.25), fontsize = 15)

#Ofankor plot
ax[1,0].plot(Ofankor_june, color = 'k', lw = 3, marker = 'o', label = 'Rainfall amount')
ax[1,0].plot(Ofankor_anon, color = 'r', lw = 3, marker = 'o', label = 'Ranifall anomaly')
ax[1,0].set_ylabel('Rainfall (mmd$^{-1}$)', fontsize =18)
ax[1,0].set_xlabel('Days', fontsize =18)
ax[1,0].set_title('Ofankor, June, 2021', size =20, weight ='bold')
ax[1,0].axhline(y = 10, color ='k', label = '75th Percentile' , linestyle = '-', lw =3)
ax[1,0].axhline(y = 25, color ='r', label = '95th Percentile', linestyle = '--', lw =3)
ax[1,0].set_xticks(range(1, 30, 3))
ax[1,0].set_xticklabels(['3','6', '9','12', '15', '18', '21','24', '27', '30'], size =20)
ax[1,0].set_yticks(range(-10, 50, 10))
ax[1,0].set_yticklabels(['-10','0','10', '20', '30', '40'], size =20)
ax[1,0].annotate('*', xy =(2.8, 2.2), size =40, color = 'b')
ax[1,0].annotate('26.26%', xy =(3, 5.7), size =30, color = 'k')
ax[1,0].legend(loc = 2, fontsize =15)

ax.flat[-1].set_visible(False) 
```
